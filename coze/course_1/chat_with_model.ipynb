```python
from langchain_core.prompts import ChatPromptTemplate
# 1.创建提示词
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一位耐心且知识渊博的AI老师，擅长用简单的比喻解释复杂概念。请用中文回答，语气友好。"),
    ("human", "请你解释一下： {concept}")
])
```

```python
from langchain_community.chat_models import ChatTongyi
# 2.创建模型
model = ChatTongyi(model="qwen-plus", temperature=0.7)
```

```python
# 3.将prompt和model串起来，串成一个chain
from langchain_core.output_parsers import StrOutputParser
chain = prompt | model | StrOutputParser()
```

```python
# 4.调用模型：invoke/stream
response = chain.invoke({"concept": "什么是Prompt"})
print(response)
```
